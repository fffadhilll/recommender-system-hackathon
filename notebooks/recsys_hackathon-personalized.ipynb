```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
def load_data(file_name):
  '''
    Menampilkan data yang tersedia.

      Parameters:
        filename (str): Nama file yang tersedia

      Return:
        File Dataframe (DataFrame)
  '''

  df = pd.read_csv(file_name)
  return df
```

```python
df = load_data('../data/amazon.csv')
df.head()
```

```python
df.info()
```

#Check Data

```python
def check_data(data):
  '''
    Mengecek data null dan data duplikat.

      Parameters:
        data: Data yang ingin dicek

      Return:
        (str): Jumlah data null dan data duplikat
  '''

  null_data = data.isnull().sum()
  duplicated_data = data.duplicated().sum()

  print(f"Null Data:\n{null_data}\n\nDuplicated Data: {duplicated_data}" )
```

```python
check_data(df)
```

#Cleaning Data

```python
df.head(1)
```

##Change price type to int

```python
df['actual_price'] = df['actual_price'].str.replace('₹', '')
df['actual_price'] = df['actual_price'].str.replace(',', '')
df['actual_price'] = df['actual_price'].str.replace('.', '')
```

```python
df['actual_price'] = df['actual_price'].astype('int')
```

##Change rating and rating_count type to float

```python
df['rating'] = pd.to_numeric(df['rating'], errors = 'coerce')
```

```python
df['rating_count'].info()
```

```python
df['rating_count'] = df['rating_count'].str.replace(',', '')
# df['rating_count'] = df['rating_count'].astype('int')
```

```python
df['rating_count'] = df['rating_count'].fillna(0)
df['rating_count'] = df['rating_count'].astype('int')
```

```python
df.head(1)
```

#EDA

```python
top_prod = df.groupby('product_id').agg({
    'rating': 'mean',
    'rating_count': 'sum'
}).sort_values(by='rating_count', ascending=False).reset_index()

top_prod = top_prod[top_prod['rating_count'] > 1000]
top_prod
```

```python
sns.scatterplot(data=top_prod, x='rating', y='rating_count')
```

```python
df_top_prod = pd.merge(df, top_prod[['product_id']], on='product_id', how='inner')
df_top_prod
```

```python
matrix = df_top_prod.pivot_table(index='user_id', columns='product_id', values='rating')
matrix
```

```python
matrix_norm = matrix.subtract(matrix.mean(axis=1), axis='rows')
matrix_norm.head()
```

```python
user_similarity = matrix_norm.T.corr()
user_similarity.head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
user_similarity_cosine = cosine_similarity(matrix_norm.fillna(0))
user_similarity_cosine
```
